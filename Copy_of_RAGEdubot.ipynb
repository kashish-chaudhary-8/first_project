```python
from pypdf import PdfReader

reader = PdfReader(pdf_file)

print("Total Pages:", len(reader.pages))
```

```python
text = ""

for page in reader.pages:
    page_text = page.extract_text()

    if page_text:
        text += page_text + "\n"

print("Characters extracted:", len(text))
```

```python
print(text[:2000])
```

```python
pages_data = []

for page_num, page in enumerate(reader.pages):

    page_text = page.extract_text()

    if page_text:
        pages_data.append({
            "page": page_num + 1,
            "text": page_text
        })

print("Pages stored:", len(pages_data))
print(pages_data[0])
```

```python
import re

def clean_text(text):

    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text)

    # Remove extra new lines
    text = text.strip()

    return text
```

```python
for page in pages_data:
    page["text"] = clean_text(page["text"])
```

```python
print(pages_data[0]["text"][:500])
```

```python
def create_chunks(text, chunk_size=500):

    chunks = []

    for i in range(0, len(text), chunk_size):

        chunk = text[i:i+chunk_size]
        chunks.append(chunk)

    return chunks
```

```python
sample_chunks = create_chunks(
    pages_data[0]["text"],
    chunk_size=500
)

print("Total Chunks:", len(sample_chunks))
print(sample_chunks[0])
```

```python
all_chunks = []
```

```python
for page in pages_data:

    page_chunks = create_chunks(
        page["text"],
        chunk_size=500
    )

    for chunk in page_chunks:

        all_chunks.append({
            "page": page["page"],
            "chunk": chunk
        })
```

```python
print("Total Chunks:", len(all_chunks))

print(all_chunks[0])
```

```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5"
)

print("Model Loaded Successfully")
```

```python
sample_text = "Machine Learning is a subset of AI."

vector = embedding_model.encode(sample_text)

print(type(vector))
print(vector.shape)
```

```python
chunk_texts = [
    item["chunk"]
    for item in all_chunks
]
```

```python
chunk_embeddings = embedding_model.encode(
    chunk_texts,
    show_progress_bar=True
)
```
